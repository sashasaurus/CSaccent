```python
import random
```

```python
#create a list of each location point 
list1 = [959, 189]
list2 = [1331, 263]
list3 = [1564, 540]
list4 = [1331, 816]
list5 = [959, 890]
list6 = [588, 816]
list7 = [355, 540]
list8 = [588, 263]
```

```python
#make one big list of all possible locations 
list_all_locations = [list1, list2, list3, list4, list5, list6, list7, list8]
```

```python
equivalent_list = [(1564, 540), (1331, 816), (960, 890), (588, 816), (355, 540), (588, 263), (959, 189), (1331, 263)]
```

```python
#use random.sample to generate a new list with two selected elements from all possible locations 
#use while loop to do this as many times needed, and populate a new list with the random selections

count = 0 
random_list = [] 

while count < 500: 
    list_of_random_locations = random.sample(list_all_locations, 2)
    random_list.append(list_of_random_locations)
    #print(list_of_random_locations)
    count = count + 1 
```

```python
#create a list of all the exceptions: neighbors, top-bottom, doubles 
exceptions = [[list1, list2], [list2, list1], [list2, list3], [list3, list2],[list3, list4], [list4, list3], [list4, list5], 
    [list5, list4], [list5, list6], [list6, list5], [list6, list7], [list7, list6], [list7, list8], [list8, list7],
    [list8, list1], [list1, list8], [list1, list5], [list5, list1], [list1, list1], [list2, list2], [list3, list3], 
    [list4, list4], [list5, list5], [list6, list6], [list7, list7], [list8, list8]]
```

```python
#for whatever is in your exceptions then remove it from the list of random locations and create a
#new list without those exceptions 
#500 input returns 345

random_with_exceptions = [x for x in random_list if x not in exceptions]

#print(random_with_exceptions)
len(random_with_exceptions)
```

```python
##Take each first index and make it into one column (target) and the second index into another column (distractor)

target_location = [item[0] for item in random_with_exceptions]

distractor_location = [item[1] for item in random_with_exceptions]
```

```python
#remove brackets and commas to make it pastable into csv 
#add parentheses around each index to conform to datasource in excel 

#remove brackets and end commas 
for item in distractor_location:
    print(str(item[0:])[1:-1])
    
```
